# Kaggle Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 # Выбор данных из большого объема данных
def get_small(path ="",size = 10, random_state=0):
    df = pd.read_csv(path)
    df.set_index([df.columns.values[0]], inplace=True)
    df.index.names = [None]
    small_benign = df.loc[df['Label'] == 'Benign'].sample(n=size,random_state=random_state)
    small_ddos = df.loc[df['Label'] == 'ddos'].sample(n=size,random_state=random_state)
    small_df = pd.concat([small_benign,small_ddos],ignore_index=True)
    small_df = small_df.sample(frac=1,random_state=random_state).reset_index(drop=True)
    return small_df
```

```python
data = get_small(path="/Users/evgenii/DDoS Dataset/final_dataset.csv",size=110000,random_state=12)
```

```python
data.head()
```

## Preprocessing

```python
# Удаляем стобцы, в которых все значения одинаковые
def unique(df):
    nunique = df.apply(pd.Series.nunique)
    cols_to_drop = nunique[nunique == 1].index
    df = df.drop(cols_to_drop, axis=1)
    return df
data = unique(data)
data
```

```python
data.info()
```

### Correlations

```python
import seaborn as sns
```

```python
features = data.columns.to_list()
target = features.pop()
```

```python
X = data.loc[:,features]
y = data.loc[:,target]
```

```python
# Поиск коррелирующих признаков
def cormap(X_data,name =""):
    plt.figure(figsize=(55,45))
    sns.heatmap(X_data.corr(), annot=True,cmap="RdBu")
    plt.title(name,fontsize=30)
    plt.plot()
    plt.savefig(name)
```

```python
cormap(X,"All correlations")
```

```python
#Удалим коррелирующие признаки при помощи фреймворка FeatureSelector
from feature_selector import FeatureSelector
```

```python
fs = FeatureSelector(data = X, labels = y)
fs.identify_collinear(correlation_threshold = 0.70)
```

```python
# список признаков для удаления
collinear_features = fs.ops['collinear']
X = X.drop(collinear_features,axis=1)
X.head()
```

```python
cormap(X,"Without correlations")
```

```python
y =y.replace("ddos",1).replace("Benign",0)
y.head()
```

```python
import time
import datetime
import ipaddress 


def toTimestamp(strtime = ""):
    timestamp =0;
    if (strtime.find('AM') == -1 and strtime.find('PM') == -1): 
        timestamp=time.mktime(datetime.datetime.strptime(strtime, "%d/%m/%Y %H:%M:%S").timetuple())
    else:
        timestamp=time.mktime(datetime.datetime.strptime(strtime, "%d/%m/%Y %H:%M:%S %p").timetuple())
    return timestamp


def iptodec(ip=""):
    return int(ipaddress.ip_address(ip))
```

```python
df_data = get_small(path="/Users/evgenii/DDoS Dataset/final_dataset.csv",size=200000,random_state=12)
df_data = unique(df_data)
X_d = df_data.loc[:,features]
y_d = df_data.loc[:,target]
X_d = X_d.drop(collinear_features,axis=1)
y_d =y_d.replace("ddos",1).replace("Benign",0)
```

## CatBoost

### With all categorial feature

```python
i
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import accuracy_score
from catboost import CatBoostClassifier, Pool
from catboost.utils import get_roc_curve
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=12)
train_data = Pool(X_train,y_train,cat_features=[0,1,3,5])
test_data = Pool(X_test,y_test,cat_features=[0,1,3,5])
```

```python
model = CatBoostClassifier(iterations=10,
                           depth=3,
                           learning_rate=0.1,
                           loss_function='Logloss',
                           verbose=False)
```

```python
model.fit(train_data)
```

Метрики для малой тестовой выборки

```python
roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
```

```python
accuracy_score(y_test, model.predict(X_test))
```

Метрики для большой тестовой выборки

```python
roc_auc_score(y_d, model.predict_proba(X_d)[:,1])
```

```python
accuracy_score(y_d, model.predict(X_d))
```
